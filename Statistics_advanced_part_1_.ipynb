#Statistics advanced part 1

#1. What is a random variable in probability theory?

A random variable is a function that assigns numerical values to each outcome in a sample space of a random experiment. It translates uncertain outcomes into quantifiable values, enabling statistical analysis. For example, in flipping a coin, a random variable X might assign 1 to heads and 0 to tails.

⸻

#2. What are the types of random variables?

There are two main types:
	•	Discrete Random Variables: These take a finite or countable number of values (e.g., number of students in a class).
	•	Continuous Random Variables: These can take any value within a range (e.g., height of individuals, time taken to complete a task).

⸻

#3. What is the difference between discrete and continuous distributions?
	•	Discrete distributions apply to random variables with distinct, separate values. Examples: Binomial, Poisson.
	•	Continuous distributions apply to variables that can take any value in a continuum. Examples: Normal, Uniform (continuous).

⸻

#4. What are probability distribution functions (PDF)?

In continuous probability, a Probability Density Function (PDF) describes the likelihood of a random variable taking on a specific value. The area under the PDF curve between two values gives the probability of the variable falling in that interval.

⸻

#5. How do cumulative distribution functions (CDF) differ from PDF?
	•	A PDF gives the density (not the probability itself) for each value in a continuous distribution.
	•	A CDF gives the actual probability that a random variable is less than or equal to a specific value, essentially accumulating the probabilities.

⸻

#6. What is a discrete uniform distribution?

This distribution assigns equal probability to all outcomes in a finite set. For example, rolling a fair die has 6 equally likely outcomes, each with a probability of 1/6.

⸻

#7. What are the key properties of a Bernoulli distribution?
	•	Models a single trial with two outcomes: success (1) or failure (0).
	•	Mean (μ) = p, Variance (σ²) = p(1 – p), where p is the probability of success.
	•	It’s the foundation for the binomial distribution.

⸻

#8. What is the binomial distribution, and how is it used in probability?

It models the number of successes in n independent Bernoulli trials, each with the same probability p. It’s used in scenarios like survey responses (e.g., 10 people saying “yes” or “no”).

⸻

#9. What is the Poisson distribution and where is it applied?

It models the number of times an event occurs in a fixed interval of time/space, assuming events occur independently and at a constant rate. Applications include call centers, traffic flow, etc.

⸻

#10. What is a continuous uniform distribution?

In this distribution, every value in a given interval [a, b] is equally likely. The PDF is flat across the range. Example: Time taken for a randomly scheduled event between 1 PM and 2 PM.

⸻

#11. What are the characteristics of a normal distribution?
	•	Symmetrical and bell-shaped.
	•	Mean = Median = Mode.
	•	Defined by its mean (μ) and standard deviation (σ).
	•	About 68% of values lie within 1σ, 95% within 2σ, and 99.7% within 3σ of the mean (Empirical Rule).

⸻

#12. What is the standard normal distribution, and why is it important?

It’s a special case of the normal distribution with μ = 0 and σ = 1. It allows for easy use of Z-tables and helps standardize variables for comparison and hypothesis testing.

⸻

#13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population’s distribution. It enables the use of normal-based inferential statistics even when the population distribution is unknown.

⸻

#14. How does the Central Limit Theorem relate to the normal distribution?

The CLT ensures that sample means are normally distributed for large enough sample sizes, making it possible to apply Z-scores and confidence intervals based on the normal distribution.

⸻

#15. What is the application of Z statistics in hypothesis testing?

Z-statistics are used when:
	•	The population standard deviation is known.
	•	Sample size is large.
They help determine how far a sample mean deviates from the population mean under the null hypothesis. It’s crucial in Z-tests.

⸻

#16. How do you calculate a Z-score, and what does it represent?

Z = (X – μ) / σ
It measures how many standard deviations a data point (X) is from the mean (μ). A positive Z means the value is above the mean; negative means below.

⸻

#17. What are point estimates and interval estimates in statistics?
	•	Point Estimate: A single value (e.g., sample mean) used to estimate a population parameter.
	•	Interval Estimate: A range (e.g., confidence interval) likely to contain the parameter, offering more reliability.

⸻

#18. What is the significance of confidence intervals in statistical analysis?

Confidence intervals provide a range of values that likely contain the population parameter. A 95% CI means we are 95% confident that the interval contains the true mean. It reflects uncertainty and precision in estimates.

⸻

#19. What is the relationship between a Z-score and a confidence interval?

Z-scores determine the critical values for confidence intervals in normal distributions. For a 95% CI, the Z-scores are ±1.96, which set the bounds around the mean.

⸻

#20. How are Z-scores used to compare different distributions?

By converting values from different distributions to Z-scores, we can compare them on a common scale. For example, comparing test scores from different exams with different means and standard deviations.

⸻

#21. What are the assumptions for applying the Central Limit Theorem?
	•	Samples are independent and identically distributed (i.i.d).
	•	Sample size is sufficiently large (typically n ≥ 30).
	•	Population should not be extremely skewed for small sample sizes.

⸻

#22. What is the concept of expected value in a probability distribution?

The expected value (E[X]) is the long-run average value of repetitions of the experiment. For discrete variables:
E[X] = Σ [x * P(x)]
It represents the theoretical mean of the distribution.

⸻

#23. How does a probability distribution relate to the expected outcome of a random variable?

A probability distribution maps all possible values of a random variable with their probabilities. The expected outcome is the average of these values weighted by their probabilities — this is the expected value.

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import randint

# Discrete uniform distribution from 1 to 6 (like a dice roll)
x = np.arange(1, 7)
pmf = randint.pmf(x, 1, 7)

plt.stem(x, pmf, use_line_collection =True)
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    x = [0, 1]
    probs = bernoulli.pmf(x, p)
    return dict(zip(x, probs))

print("Bernoulli PDF with p = 0.3:", bernoulli_pdf(0.3))
```

6. Write a Python program to calculate and plot the CDF of a discrete uniform distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values of a discrete uniform distribution
values = np.arange(1, 7)  # e.g., dice faces 1-6
probabilities = np.ones_like(values) / len(values)

# Calculate CDF
cdf = np.cumsum(probabilities)

# Plot CDF
plt.step(values, cdf, where='post', label='CDF', color='blue', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution')
plt.grid(True)
plt.legend()
plt.show()
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

# Non-normal population: exponential distribution
population = np.random.exponential(scale=2.0, size=100000)

# Parameters
sample_size = 30
num_samples = 1000
sample_means = []

# Draw samples and calculate their means
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True, color='purple', edgecolor='black', alpha=0.7)
plt.title('CLT: Sampling Distribution of Mean (Non-normal Population)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

15.Implement hypothesis testing using Z-statistics.

```python
import scipy.stats as stats

sample_mean = 52
population_mean = 50
std_dev = 4
n = 30

z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z-statistic = {z:.2f}, p-value = {p_value:.4f}")
```

16. Create confidence interval using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
sem = stats.sem(data)
confidence = 0.95
ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)

print(f"95% Confidence Interval: {ci}")
```

17.Generate normal data and calculate confidence interval of the mean.
